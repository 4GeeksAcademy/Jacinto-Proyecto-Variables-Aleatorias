# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

### DISTRIBUCIÓN NORMAL

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
# Definimos funciones de probabilidad para la distribución normal
def dnorm(x, mean=0, sd=1):
    """Calcula la estimación puntual (PDF) de la distribución normal."""
    return norm.pdf(x, loc=mean, scale=sd)
def pnorm(q, mean=0, sd=1):
    """Calcula la función de distribución acumulada (CDF) de la normal."""
    return norm.cdf(q, loc=mean, scale=sd)
def qnorm(p, mean=0, sd=1):
    """Calcula la función cuantil (PPF) de la distribución normal."""
    return norm.ppf(p, loc=mean, scale=sd)
def rnorm(n, mean=0, sd=1):
    """Genera valores aleatorios de la distribución normal."""
    return norm.rvs(size=n, loc=mean, scale=sd)
```

```python
np.random.seed(42)
mean = 0  
sd = 1    
print(f"Probability density at x=1 for Normal(0,1): {dnorm(1, mean=mean, sd=sd)}")
print(f"Probability that a Normal(0,1) is less than 1: {pnorm(1, mean=mean, sd=sd)}")
print(f"Which value in a Normal(0,1) corresponds to the 90th percentile?: {qnorm(0.9, mean=mean, sd=sd)}")
print(f"Generate 2000 random values from a Normal(0,1): {rnorm(2000, mean=mean, sd=sd)}")
```

```python
size = 200  
data = np.linspace(-5, 5, size)  
# Calculamos funciones de distribución normal
pdf = dnorm(data, mean=mean, sd=sd)  
cdf = pnorm(data, mean=mean, sd=sd)  
ppf = qnorm(np.linspace(0, 1, size), mean=mean, sd=sd)  
# Creamos la gráfica
fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].plot(data, pdf, "r-", linewidth=2)
axis[1].plot(data, cdf, "b-", linewidth=2)
axis[2].plot(data, ppf, "g-", linewidth=2)
axis[0].set_title("Función de Densidad de Probabilidad (PDF)")
axis[1].set_title("Función de Distribución Acumulada (CDF)")
axis[2].set_title("Función Cuantil (PPF)")
plt.tight_layout()
plt.show()
```

## DISTRIBUCIÓN UNIFORME

```python
# Definimos funciones de probabilidad para la distribución uniforme
def dunif(x, min_val=0, max_val=1):
    """Calcula la estimación puntual (PDF) de la distribución uniforme."""
    return uniform.pdf(x, loc=min_val, scale=max_val - min_val)

def punif(q, min_val=0, max_val=1):
    """Calcula la función de distribución acumulada (CDF) de la uniforme."""
    return uniform.cdf(q, loc=min_val, scale=max_val - min_val)

def qunif(p, min_val=0, max_val=1):
    """Calcula la función cuantil (PPF) de la distribución uniforme."""
    return uniform.ppf(p, loc=min_val, scale=max_val - min_val)

def runif(n, min_val=0, max_val=1):
    """Genera valores aleatorios de la distribución uniforme."""
    return uniform.rvs(size=n, loc=min_val, scale=max_val - min_val)
```

```python
np.random.seed(42)
min_val = 0  
max_val = 1  
print(f"Probability density at x=0.5 for Uniform(0,1): {dunif(0.5, min_val=min_val, max_val=max_val)}")
print(f"Probability that a Uniform(0,1) is less than 0.5: {punif(0.5, min_val=min_val, max_val=max_val)}")
print(f"Which value in a Uniform(0,1) corresponds to the 90th percentile?: {qunif(0.9, min_val=min_val, max_val=max_val)}")
print(f"Generate 2000 random values from a Uniform(0,1): {runif(2000, min_val=min_val, max_val=max_val)}")
```

```python
size = 200  
data = np.linspace(min_val, max_val, size)  
# Calculamos funciones de distribución uniforme
pdf = dunif(data, min_val=min_val, max_val=max_val)  
cdf = punif(data, min_val=min_val, max_val=max_val)  
ppf = qunif(np.linspace(0, 1, size), min_val=min_val, max_val=max_val)  
# Crear la gráfica
fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].plot(data, pdf, "r-", linewidth=2)
axis[1].plot(data, cdf, "b-", linewidth=2)
axis[2].plot(data, ppf, "g-", linewidth=2)
axis[0].set_title("Función de Densidad de Probabilidad (PDF)")
axis[1].set_title("Función de Distribución Acumulada (CDF)")
axis[2].set_title("Función Cuantil (PPF)")
plt.tight_layout()
plt.show()
```
